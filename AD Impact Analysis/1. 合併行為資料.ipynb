```python
def merge_data(events, dates, did_list):
    i = 1
    for date_str in dates:
        print(date_str)
        for event in events:
            path = './dataset/91APP_BehaviorData/' + event + '_' + date_str + '.csv'

            df_origin = pd.read_csv(path, encoding = "utf-8")
            
            if i == 1:
                df = df_origin.loc[df_origin['did'].isin(did_list)]
            else:
                df1 = df_origin.loc[df_origin['did'].isin(did_list)]
                df = pd.concat([df, df1])
            i += 1

    df['DateTime'] = pd.to_datetime(df.ht, unit='ms')

    df = df.sort_values(by = ['did', 'dc', 'DateTime'])
    df = df[['dc', 'did', 'uid', 'DateTime', 'bh', 'cs', 'cm', 'cn', 'pr', 'pr1id', 'pr1pr', 'pr1qt', 'cu', 'tr', 'ti']]
    
    return df
```

```python
from datetime import date, timedelta
date_range=pd.date_range(date(2020,5,1),date(2021,4,30))
date_list=[x.strftime('%F')for x in date_range]
```

```python
non_unique_uid = []
for i in range(len(group_did)):
    did = group_did[i]
    no_na = did[~pd.isna(did)]
    
    if len(no_na)>1:
        non_unique_uid.append(i)
```
