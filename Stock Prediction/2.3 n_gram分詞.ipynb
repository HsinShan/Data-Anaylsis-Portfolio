### 資料前處理
1. 移除標點符號、英文、數字
2. 依照 ngram 切割
3. 移除 stopwords


```python
def preprocess(sent,n_gram,stopwords):
    sent = re.sub(r'[^\w]',"",sent)
    sent = re.sub(r'[A-Za-z0-9]',"",sent)
    return_list = []
    for i in range(len(sent) - n_gram +1):
        w = sent[i:i+n_gram]
        if w not in stopwords:
            return_list.append(w)
    return return_list
```

### 讀取 stopwords.txt

```python
stopwords = []
with open("stopwords.TXT",'r',encoding = 'UTF8') as file:
    for data in file.readlines():
        data = data.strip()
        stopwords.append(data)
```

### 實作計算 df, tf, idf, tf-idf 的 method

```python
def calc_df(word_list):
    df = dict()
    df_doc = dict()
    for key in word_list.keys():
        for w in word_list[key]:
            if w in df:
                if key not in df_doc[w]:
                    df[w] += 1
                    df_doc[w].append(key)
            else:
                df[w] = 1
                df_doc[w] = [key] # record docID
    return df
```

```python
import math
import numpy as np
```

```python
# calculate number of times occurs in this topic
def calc_tf(word_list):
    tf = dict()
    for key in word_list.keys():
        for w in word_list[key]:
            if w in tf:
                tf[w] += 1
            else:
                tf[w] = 1
    sort_tf = {k: v for k, v in sorted(tf.items(), key=lambda item: item[1], reverse=True)}
    return sort_tf
```

```python
def calc_idf(N, df):
    idf = math.log((N/df), 10)
    return idf
```

```python
def calc_tf_idf(tf, idf):
    tf_idf = (1 + math.log(tf)) * idf
    return tf_idf
```

### 實作計算每個字的 tf-idf 並由大到小排序的 method

```python
def get_result(df_list, tf_list, N):
    result = {}

    for word in tf_list:
        idf = calc_idf(N, df_list[word])
        tf_idf = calc_tf_idf(tf_list[word], idf)
        result[word] = tf_idf
    
    sort_result = {k: v for k, v in sorted(result.items(), key=lambda item: item[1], reverse=True)}
    return sort_result
```

```python
df_list_group = {
    2: defaultdict(list),
    3: defaultdict(list),
    4: defaultdict(list),
    5: defaultdict(list),
    6: defaultdict(list)
}
# df_list_group[2] = { 1: [], 2:[]}
for i in range(2,7):
    for key, val in sorted(df_list[i].items()):
        df_list_group[i][val].append(key)
```

```python
remove_list = {2: [], 3: [], 4: [], 5: [], 6: []}
freq_list = [70, 30 ,20, 10 ,30] #為了計算快一點


for i in range(2,6):
    for word1 in df_list[i]:
        df = df_list[i][word1]
        
        if i==2 or i==3:
            value=0.06
        else:
            value=0.03
            
        upper = int(df + value * df)
        bottom = int(df - value * df)
        df_list_merge = []
 
        if df < freq_list[i-2]:
            remove_list[i].append(word1)
            continue
        
        for j in range(bottom, upper+1):
            df_list_merge += df_list_group[i+1][j]
        
        for word2 in df_list_merge:
            if word1 in word2:
                remove_list[i].append(word1)
                break  
```

### 輸出合併後tf_list和df_list

```python
import csv

def export_csv(result, file_name):
    i = 1
    with open(file_name, 'w', encoding='utf-8') as f:
        for key in result.keys():
            f.write("%s,%s\n"%(key,result[key]))
            i += 1
```

```python
def merge_list(list):
    list_merge = {
    **list[2],
    **list[3],
    **list[4],
    **list[5],
    **list[6]
    }
    sort_list = {k: v for k, v in sorted(list_merge.items(), key=lambda item: item[1], reverse=True)}
    return sort_list
```

```python
tf_list_merge = merge_list(tf_list)
export_csv(tf_list_merge, 'ngram_tf_list.csv')
```

```python
df_list_merge = merge_list(df_list)
export_csv(df_list_merge, 'ngram_df_list.csv')
```
