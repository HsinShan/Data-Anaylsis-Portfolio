```python
xls = pd.ExcelFile('dataset/stock_data.xlsx')
sheets = xls.sheet_names # see all sheet names
```

```python
data_name = ['list_2018', 'list_2017', 'list_2016', 'counter_2018', 'counter_2017', 'counter_2016']
```

```python
collections = {}
for i in range(len(sheets)):
    collections[data_name[i]] = pd.read_excel('dataset/stock_data.xlsx',index_col = 0, sheet_name = sheets[i])
```

```python
xls_2 = pd.ExcelFile('dataset2019/stock_data_2019-2020.xlsx')
sheets_v2 = xls_2.sheet_names
```

```python
sheets_v2.pop(0)
sheets_v2
```

```python
data_name_v2 = ['list_2020', 'list_2019', 'counter_2020', 'counter_2019']
```

```python
for i in range(len(sheets_v2)):
    collections[data_name_v2[i]] = pd.read_excel('dataset2019/stock_data_2019-2020.xlsx',index_col = 0, sheet_name = sheets_v2[i])
```

```python
list_data = pd.concat([
    collections['list_2020'], 
    collections['list_2019'], 
    collections['list_2018'], 
    collections['list_2017'], 
    collections['list_2016']
],axis=0)
```

```python
counter_data = pd.concat([
    collections['counter_2020'], 
    collections['counter_2019'], 
    collections['counter_2018'], 
    collections['counter_2017'], 
    collections['counter_2016']
],axis=0)
```

```python
companies = list(list_data.index.unique())
```

```python
len(companies)
```

```python
result = []

companies = list(list_data.index.unique())

for company in companies:
    stocks_data = list_data.loc[company]
#         if len(stocks_data) < 200:
#         print(i, company, len(stocks_data))
#             continue
    if stocks_data.size == 11:
        print(company, stocks_data.size, len(stocks_data))
        continue

    closing_price = stocks_data.loc[:, '收盤價(元)']

    statistic = {
        'type': '上市',
        'number': stocks_data.size / 11,
        'company': company,
        'mean': closing_price.mean(),
        'median': closing_price.median(),
        'variance': closing_price.var(),
        'std': closing_price.std()
    }
    result.append(statistic)
```
