```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
import seaborn as sns; sns.set()
```

```python
# 跑迴圈讀檔
# X_train第1~5月文章關鍵詞, X_test第6月文章關鍵詞, y_train第1~5月文章實際漲跌, y_test第6月文章實際漲跌

testing_score_list = []
pred_list = []

for i in range(30):
    index = str(i)
    
    # X_train第1~5月文章關鍵詞
    X_training_name = 'training_data/X_training_' + index + '.csv'
    Read_X_training = pd.read_csv(X_training_name, index_col = 0)
    X_train = Read_X_training.values

    # X_test第6月文章關鍵詞
    X_testing_name = 'testing_data/X_testing_' + index + '.csv'
    Read_X_testing = pd.read_csv(X_testing_name, index_col = 0)
    X_test = Read_X_testing.values

    # y_train第1~5月文章實際漲跌
    Y_training_name = 'training_data/Y_training_' + index + '.csv'
    Read_Y_training = pd.read_csv(Y_training_name)
    y_train_list = []
    for i in range(len(Read_Y_training.values)):
        num = int(Read_Y_training.values[i][0])
        y_train_list.append(num)
    y_train = np.array(y_train_list)

    # y_test第6月文章實際漲跌
    Y_testing_name = 'testing_data/Y_testing_' + index + '.csv'
    Read_Y_testing = pd.read_csv(Y_testing_name)
    y_test_list = []
    for i in range(len(Read_Y_testing.values)):
        num = int(Read_Y_testing.values[i][0])
        y_test_list.append(num)
    y_test = np.array(y_test_list)


    # 多項氏貝式分類器 Multinomial Naive Bayes
    modelm = MultinomialNB()  # alpha預設為1
    modelm.fit(X_train,y_train)  # 資料集餵入模型
    pred = modelm.predict(X_test)  # 看測試集分類結果
    for i in range(len(pred)):
        up_down = pred[i]
        pred_list.append(up_down)

    # 測試集成效
    testing_score = '%.2f' % modelm.score(X_test,y_test)
    testing_score_list.append(testing_score)

print(testing_score_list)
print(pred_list)
```

```python
result_pd = pd.DataFrame(testing_score_list) 
result_pd.to_csv(r'5+2_testing_score.csv')
```

```python
result_pd = pd.DataFrame(pred_list) 
result_pd.to_csv(r'5+2_testing_pred.csv')
```
