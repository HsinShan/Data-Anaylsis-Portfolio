## 篩選玉晶光相關新聞

```python
import pandas as pd
```

```python
data_16_18 = pd.read_csv('dataset/news.csv', encoding = "big5",engine='python')
```

```python
data_19 = pd.read_csv('dataset2019/news_2019.csv', encoding = "utf-8",engine='python', warn_bad_lines=True, error_bad_lines=False)
```

```python
data_20 = pd.read_csv('dataset2019/news_2020.csv', encoding = "utf-8",engine='python', warn_bad_lines=True, error_bad_lines=False)
```

```python
company = '玉晶光'
filter_1 = data_16_18[data_16_18['content'].str.contains(company, na=False)]
filter_2 = data_19[data_19['content'].str.contains(company, na=False)]
filter_3 = data_20[data_20['content'].str.contains(company, na=False)]
```

```python
print(len(filter_1), len(filter_2), len(filter_3))
```

```python
company_news = pd.concat([
    filter_1,
    filter_2,
    filter_3
],axis=0)
```

```python
company_news_pd = pd.DataFrame(company_news)
company_news_pd.to_csv(r'玉晶光_news.csv', index = False)
```
