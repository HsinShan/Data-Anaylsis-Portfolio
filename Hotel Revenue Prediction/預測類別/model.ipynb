```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten,LSTM,Embedding
from keras import optimizers
from keras.callbacks import EarlyStopping
```

```python
n_steps = 3
n_features = 10
model = Sequential()

# model.add(Embedding(input_dim=43, output_dim=128))
model.add(LSTM(units = 64, input_dim=42))
# model.add(LSTM(units = 64,return_sequences=True))#new
model.add(Dense(32))
model.add(Dense(16))
model.add(Dense(10, activation='sigmoid'))
model.summary()
```

```python
import numpy as np
train_lstm=np.array(train)
X_train = np.reshape(train_lstm, (train_lstm.shape[0], 1, train_lstm.shape[1]))
test_lstm=np.array(test)
X_test = np.reshape(test_lstm, (test_lstm.shape[0], 1, test_lstm.shape[1]))
y_label_tr=pd.get_dummies(y_label, columns=["label"])
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
# model.compile(loss='mean_squared_error', optimizer='adam')
hist = model.fit(X_train, y_label_tr, batch_size=20, epochs=40, verbose=1)
```

```python
test_pre_lstm=np.argmax(model.predict(X_test), axis=1)
test_pre_lstm
```

```python
test_pre_lstm=pd.DataFrame(test_pre_lstm,columns=["label"])
y_test_lstm= pd.concat([test_nolabel,test_pre_lstm],axis=1)
```
