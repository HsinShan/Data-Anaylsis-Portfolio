```python
import numpy as np
import pandas as pd
import calendar
```

```python
# train = pd.read_csv('train.csv', engine='python',index_col="ID")
test = pd.read_csv('test.csv', engine='python',index_col="ID")
# train_label = pd.read_csv('train_label.csv', engine='python')
# test_nolabel = pd.read_csv('test_nolabel.csv', engine='python')
```

```python
train=test
train
```

```python
# train_nocan=train[train["is_canceled"]!=1]
train_nocan=train
train_nocan=train_nocan.reset_index()
# train_nocan=train
```

```python
train_8=train_nocan[["arrival_date_year","arrival_date_month","arrival_date_day_of_month","meal","distribution_channel","is_repeated_guest","previous_cancellations","previous_bookings_not_canceled","reserved_room_type","assigned_room_type","booking_changes"]]
```

```python
train_8
```

```python
date=[]
month=["","January","February","March","April","May","June","July","August","September","October","November","December"]
for i in range(len(train_8)):
    mo=month.index(train_8["arrival_date_month"].iloc[i])
    if mo<10:
        mo="0"+str(mo)
    year=int(train_8["arrival_date_year"].iloc[i])
    day=int(train_8["arrival_date_day_of_month"].iloc[i])
    if int(day)<10:
        day="0"+str(day)
#     print(year)
    date.append(str(year)+"-"+str(mo)+"-"+str(day))
```

```python
len(date)
```

```python
date_pd=pd.DataFrame(date,columns=["date"])
train_8_conca = pd.concat([date_pd,train_8],axis=1)
train_8_conca
train_8_new=train_8_conca.drop(["arrival_date_year","arrival_date_month","arrival_date_day_of_month"],axis=1)
train_8_new
```

```python
train_8_meal=train_8_new[["date","meal"]]
train_8_meal
```

```python
train_8_meal=pd.get_dummies(train_8_meal, columns=["meal"])
train_8_meal_final=train_8_meal.groupby(by=["date"]).sum()
train_8_meal_final=train_8_meal_final.drop(["meal_Undefined"],axis=1)
train_8_meal_final
```

```python
train_8_disch=train_8_new[["date","distribution_channel"]]
train_8_disch
```

```python
train_8_disch=pd.get_dummies(train_8_disch, columns=["distribution_channel"])
train_8_disch_final=train_8_disch.groupby(by=["date"]).sum()
# train_8_disch_final=train_8_disch_final.drop(["distribution_channel_Undefined"],axis=1)
train_8_disch_final
```

```python
train_8_isrep=train_8_new[["date","is_repeated_guest"]]
train_8_isrep
```

```python
train_8_isrep=pd.get_dummies(train_8_isrep, columns=["is_repeated_guest"])
train_8_isrep_final=train_8_isrep.groupby(by=["date"]).sum()
# train_8_disch_final=train_8_disch_final.drop(["distribution_channel_Undefined"],axis=1)
train_8_isrep_final
```

```python
train_8_precan=train_8_new[["date","previous_cancellations"]]
train_8_precan_final=train_8_precan.groupby(by=["date"]).mean()
train_8_precan_final
```

```python
train_8_prenocan=train_8_new[["date","previous_bookings_not_canceled"]]
train_8_prenocan_final=train_8_prenocan.groupby(by=["date"]).mean()
train_8_prenocan_final
```

```python
train_8_room=train_8_new[["date","reserved_room_type","assigned_room_type"]]
train_8_room
# train_8_room_final=train_8_room.groupby(by=["date"]).mean()
# train_8_room_final
```

```python
room=[]
for i in range(len(train_8_room)):
    if train_8_room["reserved_room_type"].iloc[i]==train_8_room["assigned_room_type"].iloc[i]:
        room.append(0)
    else:
        room.append(1)
room=pd.DataFrame(room,columns=["room"])
train_8_room_conca = pd.concat([date_pd,room],axis=1)
train_8_room_final=pd.get_dummies(train_8_room_conca, columns=["room"])
train_8_room_final=train_8_room_final.groupby(by=["date"]).sum()
train_8_room_final
```

```python
train_8_assroom=train_8_new[["date","assigned_room_type"]]
train_8_assroom_du=pd.get_dummies(train_8_assroom, columns=["assigned_room_type"])
train_8_assroom_gp=train_8_assroom_du.groupby(by=["date"]).sum()
train_8_assroom_gp
```

```python
train_8_assroom_gp_ap=train_8_assroom_gp.drop(["assigned_room_type_A","assigned_room_type_D"],axis=1)
train_8_assroom_gp_ap
```

```python
train_8_assroom_gp_else=train_8_assroom_gp_ap.sum(axis=1)
train_8_assroom_gp_else=pd.DataFrame(train_8_assroom_gp_else,columns=["assigned_room_type_else"])
train_8_assroom_gp_else
```

```python
train_8_assroom_gp_final = pd.concat([train_8_assroom_gp[["assigned_room_type_A","assigned_room_type_D"]],train_8_assroom_gp_else],axis=1)
train_8_assroom_gp_final
```

```python
train_8_book=train_8_new[["date","booking_changes"]]
train_8_book_final=train_8_book.groupby(by=["date"]).mean()
train_8_book_final
```

```python
train_mid_8=pd.concat([train_8_meal_final,train_8_disch_final,train_8_isrep_final,train_8_precan_final,train_8_prenocan_final,train_8_room_final,train_8_assroom_gp_final,train_8_book_final],axis=1)
train_mid_8
```

```python
import csv

# 開啟輸出的 CSV 檔案
with open('preprocess_test_num_part2.csv', 'w', newline='') as csvfile:
    # 建立 CSV 檔寫入器
    writer = csv.writer(csvfile)

    # 寫入標題資料
    writer.writerow(["arrival_date","meal_BB","meal_FB",
                     "meal_HB","meal_SC","distribution_channel_Corporate",
                     "distribution_channel_Direct","distribution_channel_GDS",
                     "distribution_channel_TA/TO","is_repeated_guest_0",
                     "is_repeated_guest_1","previous_cancellations","previous_bookings_not_canceled","room_0","room_1",
                     "assigned_room_type_A","assigned_room_type_D","assigned_room_type_else","booking_changes"])
    # 寫入距離資料
    for i in range(len(train_mid_8)):
        writer.writerow([train_mid_8.index[i],train_mid_8["meal_BB"].iloc[i],train_mid_8["meal_FB"].iloc[i],
                         train_mid_8["meal_HB"].iloc[i],train_mid_8["meal_SC"].iloc[i],train_mid_8["distribution_channel_Corporate"].iloc[i],
                         train_mid_8["distribution_channel_Direct"].iloc[i],train_mid_8["distribution_channel_GDS"].iloc[i],
                         train_mid_8["distribution_channel_TA/TO"].iloc[i],train_mid_8["is_repeated_guest_0"].iloc[i],
                         train_mid_8["is_repeated_guest_1"].iloc[i],train_mid_8["previous_cancellations"].iloc[i],
                         train_mid_8["previous_bookings_not_canceled"].iloc[i],train_mid_8["room_0"].iloc[i],
                         train_mid_8["room_1"].iloc[i],train_mid_8["assigned_room_type_A"].iloc[i],train_mid_8["assigned_room_type_D"].iloc[i],
                        train_mid_8["assigned_room_type_else"].iloc[i],train_mid_8["booking_changes"].iloc[i]])
```

```python
quatity=pd.DataFrame(train_8_book.groupby(by=["date"]).size(),columns=["quatity"])
quatity
```

```python
train_mid_8_pe=pd.concat([train_mid_8,quatity],axis=1)
train_mid_8_pe
```
