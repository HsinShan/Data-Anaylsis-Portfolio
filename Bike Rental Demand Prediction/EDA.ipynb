# 用迴歸解釋

```python
import statsmodels.api as sm
from sklearn import linear_model

# x與y變數'
y = data['casual']
x = data.copy()
x = x.drop(['instant','dteday', 'casual','registered','cnt'], axis=1)  # , 'weathersit'

# LinearRegression model
X2 = sm.add_constant(x)    # 為模型增加常數項，即迴歸線在y軸上的截距

models = sm.OLS(y, X2).fit()
print("未註冊的使用者的當日總租賃數量")
print(models.summary())
```

# 在註冊會員與非註冊使用者，season季節與temp溫度關係

```python
# 非註冊使用者
df1 = data[data['season']==1]
df2 = data[data['season']==2]
df3 = data[data['season']==3]
df4 = data[data['season']==4]

plt.scatter("temp", "casual", data=df1, label = 1)
plt.scatter("temp", "casual", data=df2, label = 2)
plt.scatter("temp", "casual", data=df3, label = 3)
plt.scatter("temp", "casual", data=df4, label = 4)
plt.legend()
plt.show()
```

```python
# 註冊會員
df1 = data[data['season']==1]
df2 = data[data['season']==2]
df3 = data[data['season']==3]
df4 = data[data['season']==4]

plt.scatter("temp", "registered", data=df1, label = 1)
plt.scatter("temp", "registered", data=df2, label = 2)
plt.scatter("temp", "registered", data=df3, label = 3)
plt.scatter("temp", "registered", data=df4, label = 4)
plt.legend()
plt.show()
```
